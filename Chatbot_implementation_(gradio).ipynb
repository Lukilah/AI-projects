```python
import gradio as gr
import transformers
import torch
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
model = AutoModelForCausalLM.from_pretrained("AlvinxLukilah/UpdatedTextGen")
# Use the GPT-2 tokenizer
tokenizer = AutoTokenizer.from_pretrained("AlvinxLukilah/UpdatedTextGen")
```

```python
# Keep track of previous responses
previous_responses = []
```

```python
# Define a function for the chatbot
def chatbot(input_text):
    global previous_responses
    # Tokenize the user's input
    input_ids = tokenizer.encode(input_text, return_tensors="pt")
    # Create an attention mask
    attention_mask = torch.ones(input_ids.shape)  # Set all tokens to 1 to indicate "pay attention"

    # Generate a response from the model
    response = model.generate(input_ids, attention_mask=attention_mask, max_length=180, num_return_sequences=1)

    # Decode and return the response
    return tokenizer.decode(response[0], skip_special_tokens=True)

    # Check for repetition in the generated text
    if any(prev_response in generated_text for prev_response in previous_responses):
        return "Chatbot: I'm sorry, but I'm having difficulty generating a unique response."
    # Add the generated text to the list of previous responses
    previous_responses.append(generated_text)

    return "Chatbot: " + generated_text
```

```python
# Define a function to handle user input and get a response
def get_chatbot_response(prompt):
    return chatbot(prompt)
```

```python
iface = gr.Interface(
    fn=get_chatbot_response,
    inputs=gr.Textbox(),
    outputs="text",
    #live=True,
    title="Mental Health Chatbot",
    description="Hi🌱, enter a message and the chatbot will respond accordingly :)",
)
```

```python
# Launch the Gradio interface
iface.launch()
```
